```python
import pandas as pd
import mysql.connector

def run_query(query):
    # Connect to the database
    cnx = mysql.connector.connect(user='root', password='qtr2rp8x',
                                  host='localhost', database='gene_blast')

    # Create a cursor object
    cursor = cnx.cursor()

    # Execute the query
    cursor.execute(query)

    # Fetch all the data from the cursor
    rows = cursor.fetchall()

    # Get the column names
    column_names = [i[0] for i in cursor.description]

    # Convert the data into a pandas DataFrame with the header
    df = pd.DataFrame(rows, columns=column_names)

    # Close the cursor and connection
    cursor.close()
    cnx.close()

    return df
```

```python
# Query 1: Simple SELECT query
query1 = "SELECT * FROM classification_database"
df_1 = run_query(query1)

df_1
```

```python
# Query 2: SELECT query with LIMIT
query2 = "SELECT * FROM classification_database LIMIT 10"
df2 = run_query(query2)
df2
```

```python
# Query 3: SELECT query with WHERE
query3 = "SELECT * FROM classification_database WHERE score > 50"
df3 = run_query(query3)
df3.head()
```

```python
# Query 4: SELECT query with ORDER BY
query4 = "SELECT * FROM classification_database ORDER BY score DESC"
df4 = run_query(query4)
df4.head()
```

```python
# Query 5: SELECT query with GROUP BY
query5 = "SELECT symptom, COUNT(*) as count FROM classification_database GROUP BY symptom"
df5 = run_query(query5)
df5
```

```python
# Query 6: SELECT query with HAVING
query6 = "SELECT symptom, COUNT(*) as count FROM classification_database GROUP BY symptom HAVING count > 10"
df6 = run_query(query6)
df6
```

```python
# Query 7: SELECT query with INNER JOIN
query7 = """
SELECT c1.id, c1.score, c1.symptom, c2.user_type
FROM classification_database c1
INNER JOIN df1 c2 ON c1.id = c2.id
"""

df7 = run_query(query7)
df7.head()
```

```python
# Query 8 SELECT query with LEFT JOIN
query8 = """
SELECT c1.id, c1.score, c1.symptom, c2.user_type
FROM classification_database c1
LEFT JOIN df1 c2 ON c1.id = c2.id
"""

df8 = run_query(query8)
df8.head()
```

```python
# Query 9 SELECT query with subquery
query9 = """
SELECT id, score, symptom
FROM classification_database
WHERE id IN (SELECT id FROM df1 WHERE user_type = 'a')
"""

df9 = run_query(query9)
df9
```
